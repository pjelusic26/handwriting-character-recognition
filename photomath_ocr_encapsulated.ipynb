```python
# Importing class
from photomath_script import Photomath

# Defining new line
nl = '\n'
```

```python
# Reading and processing image
img_orig = 'testset/2plus2.jpg'
img_processed = Photomath.imgPreprocess(img_orig)
```

```python
# Displaying image
Photomath.showImage(img_processed, 9, f"")
```

```python
text, data = Photomath.recognizeCharacters(img_processed)
```

```python
digits, operators = Photomath.divideString(text)

# Printing digits and operators
print(f"Digits:{nl}{digits}")
print(f"Operators:{nl}{operators}")

# Calculating result
result = Photomath.result(digits, operators)

# Printing result
print(f"Result:{nl}{result}")
```

```python
# Saving recognized characters
Photomath.saveCharacters(img_processed, img_orig)
```
