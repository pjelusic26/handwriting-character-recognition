```python
# Importing classes
from photomath_script import Photomath
```

```python
# Running the model, with only 10000 train samples - better results + faster time
model = Photomath.fitMnist(10000)
```

```python
# Importing 3 custom images
images = Photomath.importImages('testset/invert_four.png', 'testset/invert_two.png', 'testset/invert_zero.png')
```

```python
# Making predictions
predictions = Photomath.getPredictions(model, images)
```

```python
# Displaying custom images with their corresponding predictions

for index, value in enumerate(predictions):
    Photomath.showImage(images[index], 2, f"Prediction: {value}")
```
